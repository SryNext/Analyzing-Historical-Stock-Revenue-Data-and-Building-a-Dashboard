```python
import yfinance as yf

# Download Tesla stock data
tesla_stock = yf.Ticker("TSLA")
tesla_data = tesla_stock.history(period="max")

# Reset the index
tesla_data.reset_index(inplace=True)

# Save the data to a CSV file
tesla_data.to_csv('tesla_data.csv', index=False)

# Display the first five rows
print(tesla_data.head())
```

```python
import requests
from bs4 import BeautifulSoup
import pandas as pd

# URL to extract Tesla revenue data
url = 'https://www.macrotrends.net/stocks/charts/TSLA/tesla/revenue'

# Send a request to the website
response = requests.get(url)

# Create a BeautifulSoup object and parse the HTML content
soup = BeautifulSoup(response.text, 'html.parser')

# Find the table containing the revenue data
table = soup.find_all('table', {'class': 'historical_data_table table'})[0]

# Extract table headers and rows
headers = [header.text for header in table.find_all('th')]
rows = table.find_all('tr')[1:]

# Extract the data and create a list of dictionaries
data = []
for row in rows:
    cells = row.find_all('td')
    row_data = {headers[i]: cells[i].text for i in range(len(headers))}
    data.append(row_data)

# Create a DataFrame from the list of dictionaries
tesla_revenue = pd.DataFrame(data)

# Display the last five rows
print(tesla_revenue.tail())
```

Question 5: Plot Tesla Stock Graph

```python
import matplotlib.pyplot as plt

def make_graph(stock_data, title):
    plt.figure(figsize=(10, 5))
    plt.plot(stock_data['Date'], stock_data['Close'], label='Close')
    plt.xlabel('Date')
    plt.ylabel('Close Price')
    plt.title(title)
    plt.legend()
    plt.show()

# Assuming you have already downloaded the Tesla stock data using yfinance
# and saved it in the tesla_data DataFrame (see code from Question 1),
# you can now plot the graph using the make_graph function:
make_graph(tesla_data, 'Tesla Stock Data')
```
